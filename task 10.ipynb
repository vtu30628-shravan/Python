```python
# Program: Visualize Student Performance Using Matplotlib

import matplotlib.pyplot as plt

# Student names and marks
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eva']
math = [85, 90, 78, 92, 88]
science = [90, 82, 85, 80, 95]
english = [88, 96, 89, 90, 84]

# Line Chart
plt.figure(figsize=(8,4))
plt.plot(students, math, marker='o', label='Math')
plt.plot(students, english, marker='x', label='English')
plt.plot(students, science, marker='s', label='Science')
plt.title("Student Performance in Line Chart")
plt.xlabel('Students')
plt.ylabel('Marks')
plt.legend()
plt.grid(True)
plt.show()

# Bar Chart
plt.figure(figsize=(8,4))
x = range(len(students))
plt.bar(x, math, width=0.25, label='Math')
plt.bar([i + 0.25 for i in x], science, width=0.25, label='Science')
plt.bar([i + 0.50 for i in x], english, width=0.25, label='English')
plt.xticks([i + 0.25 for i in x], students)
plt.title("Student Marks Bar Chart")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.legend()
plt.show()

# Pie Chart for a single student (example: Charlie)
subjects = ['Math', 'Science', 'English']
charlie_marks = [math[2], science[2], english[2]]

plt.figure(figsize=(5,5))
plt.pie(charlie_marks, labels=subjects, autopct='%1.1f%%')
plt.title("Charlies Subject-wise Performance (Pie Chart)")
plt.show()
```
